```python
# 3.2 Get target and predictors:
y = mobile_price.pop("mobile_wt")
X = mobile_price
```

```python
# 4.0 Build model:
model = tf.keras.Sequential()
```

```python
# 4.1 First/Input layer
model.add(layers.Input( shape = (X.shape[1],) ) )
```

```python
# Adding Hidden Layers with 1000 Neurons
model.add(layers.Dense(1000,activation='relu'))
```

```python
# Adding Hidden Layers with 400 Neurons
model.add(layers.Dense(400,activation='relu'))
```

```python
# Adding Hidden Layers with 200 Neurons
model.add(layers.Dense(200,activation='relu'))
```

```python
#4.3 Compile, measuring the error using Mean Absolute Error
model.compile( loss = "mae")
```

```python
#4.3 Compile, measuring the error using Mean Squared Error
model.compile(loss="mse")
```

```python
model.fit(X,y, epochs = 500 )
```

```python
# FOR MAE:
# At the End of 500 EPOCHS, When there are 3 hidden layers with neurons 1000, 400 & 200 respectively, there is minimum test loss of 30.0348 at 447th Epoch.

# FOR MSE:
# At the End of 500 EPOCHS, When there are 3 hidden layers with neurons 1000, 400 & 200 respectively, there is minimum test loss of 1391.9116 at 499th Epoch. }
```
